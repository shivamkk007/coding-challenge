```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_nodes_from([2,3,4,5,6])
G.add_edge(1,2)
G.add_edges_from([(2,3),(2,5),(3,4),(3,5),(2,4),(5,6),(1,6),(1,5)])
```

```python
nx.draw_networkx(G, with_labels=True,node_color='orange',font_weight='roman',node_size=200)
plt.show()
```
